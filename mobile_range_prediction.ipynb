```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
```

```python
#train_test_split

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state = 24)
```

Support Vector Machine

```python
from sklearn import metrics
svc = SVC()
svc.fit(x_train,y_train)
svc_pred = svc.predict(x_test)
svc_conf_matrix = confusion_matrix(y_test, svc_pred)
svc_acc_score = accuracy_score(y_test, svc_pred)
print('confusion_matrix:\n\n',svc_conf_matrix,'\n')
print('SVC_accuracy_score:',svc_acc_score,'\n')
print('classification_report: \n',classification_report(y_test,svc_pred))
```

Naive Bayes

```python
nb = GaussianNB()
nb.fit(x_train,y_train)
nbpred = nb.predict(x_test)
nb_conf_matrix = confusion_matrix(y_test, nbpred)
nb_acc_score = accuracy_score(y_test, nbpred)
print('confusion_matrix:\n\n',nb_conf_matrix,'\n')
print('NB_accuracy_score:',nb_acc_score,'\n')
print('classification_report: \n',classification_report(y_test,nbpred))
```

Random Forest
